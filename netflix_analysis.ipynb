## Handling Missing Values

```python
df1 = dataset.copy()
```

```python
# For Duration column there are 3 null values that will be fixed manually
dataset[dataset['duration'].isnull()]
```

```python

# Correcting assignment syntax
df1.loc[(df1['duration'].isna()) & (df1['show_id'] == 's5542'), 'duration'] = '74 min'
df1.loc[(df1['duration'].isna()) & (df1['show_id'] == 's5795'), 'duration'] = '84 min'
df1.loc[(df1['duration'].isna()) & (df1['show_id'] == 's5814'), 'duration'] = '66 min'

# Check remaining NaN values in the 'duration' column
print(df1['duration'].isnull().sum())
```

```python
# Now for rating column there are 4 null values and some irrelevent values
dataset[dataset['rating'].isnull()]
```

```python
# Correcting assignment syntax
df1.loc[(df1['rating'].isna()) & (df1['show_id'] == 's5990'), 'rating'] = 'U/A 13+'
df1.loc[(df1['rating'].isna()) & (df1['show_id'] == 's6828'), 'rating'] = 'TV-14'
df1.loc[(df1['rating'].isna()) & (df1['show_id'] == 's7313'), 'rating'] = 'TV-G'
df1.loc[(df1['rating'].isna()) & (df1['show_id'] == 's7538'), 'rating'] = 'PG-13'

# Check remaining NaN values in the 'duration' column
print(df1['rating'].isnull().sum())
```

```python
df1['rating'].unique()
```

```python
valid_ratings = ['PG-13', 'TV-MA', 'PG', 'TV-14', 'TV-PG', 'TV-Y', 'TV-Y7', 'R',
                 'TV-G', 'G', 'NC-17', 'NR', 'U/A 13+', 'TV-Y7-FV', 'UR']
invalid_ratings = df1[~df1['rating'].isin(valid_ratings)]
invalid_ratings
```

```python
df1['rating'] = df1['rating'].replace({'74 min': 'TV-MA', '84 min': 'TV-MA', '66 min': 'TV-MA'})
```

```python
#To analyze how content distribution has changed over the years, we need date added column but it has some data type problems so i was not able to fix it
```

```python
#Top 10 directors with the most Movies/TV Shows

temp_df4 = df2.copy()

temp_df4 = temp_df4.dropna(subset=["director"])

# Split multiple directors and expand into multiple rows
temp_df4["director"] = temp_df4["director"].str.split(", ")
temp_df4 = temp_df4.explode("director")

# Count the number of titles per director
director_counts = temp_df4["director"].value_counts().head(10)

# Plot the top 10 directors
plt.figure(figsize=(12, 6))
director_counts.sort_values().plot(kind="barh", color="purple")
plt.xlabel("Number of Titles")
plt.ylabel("Director")
plt.title("Top 10 Directors with Most Movies/TV Shows on Netflix")
plt.grid(axis="x")
```

```python
# Number of Movies vs. TV Shows per Year
content_per_year = df2.groupby(["release_year", "type"]).size().unstack()

plt.figure(figsize=(12, 6))
plt.plot(content_per_year.index, content_per_year["Movie"], marker="o", label="Movies", color="blue")
plt.plot(content_per_year.index, content_per_year["TV Show"], marker="o", label="TV Shows", color="red")
plt.xlabel("Release Year")
plt.ylabel("Number of Titles")
plt.title("Movies vs. TV Shows Released Each Year")
plt.legend()
plt.grid()
plt.show()
```

```python
# Top 5 Countries Producing the Most Content
temp_df5 = df2.copy()
temp_df5 = temp_df5.dropna(subset=["country"])

temp_df5["country"] = temp_df5["country"].str.split(", ")
dtemp_df5 = temp_df5.explode("country")

top_countries = temp_df5["country"].value_counts().head(5)

plt.figure(figsize=(10, 6))
top_countries.sort_values().plot(kind="barh", color="green")
plt.xlabel("Number of Titles")
plt.ylabel("Country")
plt.title("Top 5 Countries Producing the Most Content on Netflix")
plt.grid(axis="x")
plt.show()
```

# Data visualization

##  Content Distribution Over Time (Line Chart)

```python
plt.figure(figsize=(12, 6))
plt.plot(content_per_year.index, content_per_year["Movie"], marker="o", label="Movies", color="blue")
plt.plot(content_per_year.index, content_per_year["TV Show"], marker="o", label="TV Shows", color="red")
plt.xlabel("Release Year")
plt.ylabel("Number of Titles")
plt.title("Movies vs. TV Shows Released Each Year")
plt.legend()
plt.grid()
plt.show()
# Insight: This chart shows whether Netflix has been adding more Movies or TV Shows over time.
```

##  Duration vs. Release Year (Scatter Plot for Movies Only)

```python
temp_df7 = df2.copy()
movies = temp_df7[temp_df7["type"] == "Movie"]
movies = movies[["title", "release_year", "duration", "listed_in", "country"]]
movies["duration"] = movies["duration"].str.replace(" min", "").astype(float)

avg_duration = movies["duration"].mean()
print(f"Average Movie Duration: {avg_duration:.2f} minutes")
```

```python
plt.figure(figsize=(12, 6))
plt.scatter(movies["release_year"], movies["duration"], alpha=0.5, color="blue")
plt.xlabel("Release Year")
plt.ylabel("Duration (minutes)")
plt.title("Movie Duration Over the Years")
plt.grid()
plt.show()
# Insight: This helps analyze if movie lengths have increased or decreased over time.
```
